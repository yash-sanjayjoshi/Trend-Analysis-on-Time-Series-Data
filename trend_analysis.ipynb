## 5. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Libraries used:

Pandas → Data manipulation

NumPy → Numerical calculations

Matplotlib → Graph visualization

Seaborn → Advanced visualization

## 6. Load Dataset

```python
data = pd.read_csv("monthly_sales.csv")
data.head()
```

This step loads the dataset and displays the first few records.

## 7. Data Preprocessing

Convert the date column into datetime format and set it as index.

```python
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)
```

This allows us to perform time-series analysis using dates.

## 8. Visualizing Sales Trend

```python
plt.figure(figsize=(10,5))
plt.plot(data.index, data['sales'])
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

This graph shows how sales change over time.

Observation:

1.Sales generally increase over time

2.Some fluctuations occur between months

## 9. Moving Average (Trend Smoothing)

Moving average helps remove short-term noise and highlight long-term trends.

```python
data['moving_avg'] = data['sales'].rolling(window=3).mean()

plt.figure(figsize=(10,5))
plt.plot(data.index, data['sales'], label='Original Sales')
plt.plot(data.index, data['moving_avg'], label='Moving Average')
plt.legend()
plt.show()
```

Benefits:

1.Smooths the graph

2.Highlights real sales trends

## 10. Yearly Growth Analysis

```python
data['year'] = data.index.year
yearly_sales = data.groupby('year')['sales'].sum()

yearly_sales.plot(kind='bar')
plt.title("Yearly Sales Comparison")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

This step compares sales performance across years.

## 11. Monthly Seasonality Analysis

```python
data['month'] = data.index.month
monthly_pattern = data.groupby('month')['sales'].mean()

monthly_pattern.plot()
plt.title("Monthly Seasonality Pattern")
plt.xlabel("Month")
plt.ylabel("Average Sales")
plt.show()
```

This helps identify seasonal patterns in sales.

Example:

1.Sales may increase during festival seasons

2.Sales may decrease during off-season months
